```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Import data
ords_prods_cust_merge_no_PII = pd.read_pickle(r"C:\Users\User\Documents\07-2022 Instacart Basket Analysis\02 Data\Prepared Data\ords_prods_cust_merge_no_PII.pkl")
```

```python
# Define function to create marrital and dependent tag
def familial(row):

   if row['fam_status'] == 'married' and row['n_dependants'] == 0:
     return 'Married with no kids'
   elif row['fam_status'] == 'married' and row['n_dependants'] >0:
     return 'Married with kids'
   else: return 'Not married'
```

```python
# Use function to create column
ords_prods_cust_merge_no_PII['familial_status'] = ords_prods_cust_merge_no_PII.apply(familial, axis=1)
```

```python
# View new column
ords_prods_cust_merge_no_PII.head()
```

```python
# Show distribution of familial status
ords_prods_cust_merge_no_PII['familial_status'].value_counts().plot.bar()
```

```python
ords_prods_cust_merge_no_PII.loc[ords_prods_cust_merge_no_PII['prices'] > 100]
```

```python
ords_prods_cust_merge_no_PII.loc[ords_prods_cust_merge_no_PII['prices'] >100, 'prices'] = np.nan
```

```python
ords_prods_cust_merge_no_PII['prices'].max()
```

```python
ords_prods_cust_merge_no_PII.groupby('familial_status').agg({'days_since_prior_order': ['mean', 'min', 'max']})
```

```python
Freq = ords_prods_cust_merge_no_PII.groupby('familial_status').agg({'days_since_prior_order': ['mean']})
Freq
```

```python
Freq_bar = Freq.plot.bar(title = 'Order Frequency by Familial Status')
plt.legend().remove()
```

```python
ords_prods_cust_merge_no_PII.groupby('familial_status').agg({'prices': ['mean', 'min', 'max']})
```

```python
Expend = ords_prods_cust_merge_no_PII.groupby('familial_status').agg({'prices': ['mean']})
Expend
```

```python
Expend_bar = Expend.plot.bar(title = 'Prices by Familial Status')
plt.legend().remove()
```
