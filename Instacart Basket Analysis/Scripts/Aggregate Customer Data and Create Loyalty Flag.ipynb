```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import data
ords_prods_merge = pd.read_pickle(r"C:\Users\User\Documents\07-2022 Instacart Basket Analysis\02 Data\Prepared Data\orders_products_merged.pkl")
```

```python
# Find aggregated mean of order_number for entire data frame
```

```python
# Group by product name
ords_prods_merge.groupby('product_name')
```

```python
# View mean values for entire data frame
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

#### Analyze the result -- for the most part, the subset was a pretty accurate representation of the entire population

```python
# Create a loyalty flag for existing customers
```

```python
# Group by user id
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Create flag 'loyal customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Create flag 'regular customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Create flag 'new customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# show price average by loyalty flag
```

```python
# show price average by loyalty flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
# Create flag based on spending habits
```

```python
# Create column for average price
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.average)
```

```python
# View new column
ords_prods_merge.head()
```

```python
# Create flag 'low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Create flag 'high spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Print frequency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Create flag based on frequency
```

```python
# Create column for median frequency
ords_prods_merge['med_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# View new column
ords_prods_merge.head()
```

```python
# Create flag 'non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['med_freq'] > 20, 'freq_flag'] = 'Non-frequent customer'
```

```python
# Create flag 'regular customer'
ords_prods_merge.loc[(ords_prods_merge['med_freq'] <= 20) & (ords_prods_merge['med_freq'] > 10), 'freq_flag'] = 'Regular customer'
```

```python
# Create flag 'frequent customer'
ords_prods_merge.loc[ords_prods_merge['med_freq'] <= 10, 'freq_flag'] = 'frequent customer'
```
